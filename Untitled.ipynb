```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from config import api_key
```

```python
from citipy import citipy
```

```python
# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []
countries = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    country = citipy.nearest_city(lat_lng[0], lat_lng[1]).country_code 
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)
        countries.append(country)
# Print the city count to confirm sufficient count
len(cities)
```

```python
city_data = []
countries = []
cloud = []
date = []
humid = []
lat = []
lng = []
temp = []
wind = []


for city in cities:


    url = "http://api.openweathermap.org/data/2.5/weather?"
    data_url = url + "appid="+ api_key + f"&q={city}&units=Imperial"
    weather_data = requests.get(data_url).json()

    #print(weather_data)

    try:
        cloud.append(weather_data["clouds"]["all"])
        date.append(weather_data["dt"])
        humid.append(weather_data["main"]["humidity"])
        lat.append(weather_data["coord"]["lat"])
        lng.append(weather_data["coord"]["lon"])
        temp.append(weather_data["main"]["temp_max"])
        wind.append(weather_data["wind"]["speed"])


        print(f"Processing {city}")

    except KeyError:
        print("City not found. Skipping...")
```

```python
base_data = pd.DataFrame({"City": city,
                          #"Country": countries,
                          "Lat": lat,
                          "Lng": lng,
                          "Cloudiness": cloud,
                          "Humidity": humid,
                          "Max Temp": temp,
                          "Wind Speed": wind,
                          "Date": date})

base_data.head()
```

```python
#Lat vs Temp
plt.scatter(base_data["Lat"], base_data["Max Temp"], marker="o", facecolors="b",edgecolors="black")

plt.title("City Latitude vs Max Temperature(7/31/2019)")
plt.xlabel("Latitude")
plt.ylabel("Max Temperature(F)")
plt.grid()

plt.show()
```

```python
#Lat vs Cloudiness
plt.scatter(base_data["Lat"], base_data["Humidity"], marker="o", facecolors="b",edgecolors="black")

plt.title("City Latitude vs Cloudiness(7/31/2019)")
plt.xlabel("Latitude")
plt.ylabel("Clodiness(%)")
plt.grid()

plt.show()
```

```python
#Lat vs Wind Speed
plt.scatter(base_data["Lat"], base_data["Wind Speed"], marker="o", facecolors="b",edgecolors="black")

plt.title("City Latitude vs Wind Speed(7/31/2019)")
plt.xlabel("Latitude")
plt.ylabel("Wind Speed(mph)")
plt.grid()

plt.show()
```
